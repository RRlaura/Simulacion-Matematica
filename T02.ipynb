1. Graficar las siguientes funciones f(x). Determina el dominio a partir de la función:

$  x^2-4x+2 $


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = x**2 - 4*x + 2
plt.figure()
plt.plot(x, y)
```

```python
import sympy as sp
x = sp.symbols('x')
f1 = x**2 - 4*x + 2
dominio = sp.solve(f1, x)
print(f"{dominio}")
```

2. Graficar los siguientes grupos de funciones


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y1 = 2*x+1
y2 = -x + 4
y3 = x**2
plt.plot(x, y1, label='y1 = 2*x+1')
plt.plot(x, y2, label='y2 = -x+4')
plt.plot(x, y3, label='y3 = x**2')

plt.title('Grupo 1')
plt.figure()
```

3. Crea una función que cree una gráfica y dibuje 'k' líneas de f(x)= sin(ax). La función recibe:
* x1: valor inicial de 'x'* 
x2: valor final de 'x
* n: cantidad de puntos del vector 'x'
* a: valor inicial de a
* b: valor final de a
* k: cantidad de lineasíneas

```python
import numpy as np
import matplotlib.pyplot as plt

def lineas(x1, x2, n, a, b, k):

    x = np.linspace(x1, x2, n)
    valoresdea = np.linspace(a, b, k)

    plt.figure(figsize=(10, 5))
    
    for i, ai in enumerate(valoresdea): #esto es para graficar cada linea
        y = np.sin(ai * x)
        plt.plot(x, y, label=f'a = {ai:.2f}') #esto es para usar el label en la leyenda

    plt.legend() #la leyenda
    plt.show()

lineas(x1=0, x2=2*np.pi, n=1000, a=1, b=3, k=3)
```
